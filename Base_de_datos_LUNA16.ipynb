```python
import os
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
folder = "/content/drive/MyDrive/2025/Mineria-de-datos/subset1"
```

```python
raw_files = [f for f in os.listdir(folder) if f.lower().endswith(".raw")]

if not raw_files:
    raise Exception("No se encontraron archivos .raw en la carpeta.")
```

```python
def intentar_formas(data):
    posibles = []
    dims_comunes = [(512, 512), (256, 256), (128, 128)]
    N = len(data)

    for w, h in dims_comunes:
        if N % (w*h) == 0:
            d = N // (w*h)
            posibles.append((d, h, w))

    return posibles
```

```python
dropdown = widgets.Dropdown(
    options=raw_files,
    description="Archivo RAW:",
    value=raw_files[0]
)

display(dropdown)

def visualizar_raw(nombre_archivo):
    ruta = os.path.join(folder, nombre_archivo)
    print(f"\nCargando archivo: {ruta}")

    data = np.fromfile(ruta, dtype=np.int16)
    print(f"Tamaño total: {len(data)} valores")

    formas = intentar_formas(data)

    if not formas:
        print("❌ No se encontró forma 3D compatible.")
        return

    print("Formas posibles detectadas:")
    for f in formas:
        print(f"  {f}")

    d, h, w = formas[0]
    volumen = data.reshape((d, h, w))

    print(f"Mostrando volumen con forma {volumen.shape}")

    paso = max(1, d // 20)

    fig, axes = plt.subplots(3, 4, figsize=(15, 10))
    axes = axes.flatten()

    idx = 0
    for i in range(0, d, paso):
        if idx >= 12:
            break
        axes[idx].imshow(volumen[i], cmap='gray')
        axes[idx].set_title(f"Slice {i}")
        axes[idx].axis("off")
        idx += 1

    plt.tight_layout()
    plt.show()

button = widgets.Button(description="Visualizar")
button.on_click(lambda b: visualizar_raw(dropdown.value))
display(button)
```
